## Introdução - O que é Deep Learning?

O termo Deep Learning é traduzido como “aprendizado profundo”. 

A ideia é otimizar ainda mais o aprendizado de máquina, com técnicas ainda mais avançadas, reforçando o desenvolvimento e a evolução dos sistemas artificiais. 

O Deep Learning (DL) é uma subárea do Machine Learning (ML) que, por sua vez, é uma parte da inteligência artificial.

O objetivo é utilizar um raciocínio semelhante ao que os seres humanos utilizam para aprender e identificar padrões.

Aprendizado profundo funciona com base em redes neurais, que simulam as conexões entre neurônios humanos.

Eles trabalham juntos e cada um complementa a informação do vizinho. 

Desse modo, o algoritmo consegue realizar determinadas funções e ir de um ponto A a um ponto B, sem intervenção humana.

Com isso, é possível executar um conjunto de passos para resolver um problema sem a necessidade de programação humana. 

A ideia do DL é lidar com questões mais complexas e que envolvem mais dados do que ML.

Fonte: https://blog.smartconsulting.com.br/deep-learning/

## O que são Redes Neurais?

São sistemas artificiais inspirados no cérebro humano, por isso “neurais”. 

Elas têm como objetivo replicar, nas máquinas, a maneira como as pessoas aprendem.

As redes neurais permitem que máquinas consigam entender e aprender padrões, adaptar-se a mudanças e imitar processos de pensamento humano na busca por respostas e soluções. 

Tudo isso com quase nenhuma intervenção humana.

Fonte: https://www.pollux.com.br/blog/redes-neurais-o-que-sao-e-como-usa-las-na-industria/

## Funcionamento

Funciona com uma estrutura de rede de neurônios artificiais, como já falamos. 

Essa rede se divide em camadas, com vários nós que raciocinam e tiram conclusões sobre os dados. 

Cada camada está em um nível de inteligência que permite compreender algum aspecto diferente acerca do objeto que está sendo analisado. 

As camadas cooperam entre si, partindo das conclusões das vizinhas.

## Aplicação 

Com esse raciocínio, os algoritmos inteligentes são aplicáveis em diversos segmentos e áreas. 

Um deles é o estudo das preferências e comportamentos do cliente, o que pode ser feito com base em dados históricos sobre suas escolhas de acesso e compras.

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
sess = tf.InteractiveSession()
```

```python
new_saver = tf.train.import_meta_graph('modelo/model.ckpt-900.meta')
new_saver.restore(sess, 'modelo/model.ckpt-900')
tf.get_default_graph().as_graph_def()
```

```python
x = sess.graph.get_tensor_by_name("input:0")
y_conv = sess.graph.get_tensor_by_name("output:0")
```

```python
img = mpimg.imread('images_teste/image05.jpg')     
gray = rgb2gray(img)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
image_0 = np.resize(gray,(1,48,48,1))
tResult = testResult()
num_evaluations = 50
```

```python
for i in range(0, num_evaluations):
    result = sess.run(y_conv, feed_dict={x:image_0})
    label = sess.run(tf.argmax(result, 1))
    label = label[0]
    label = int(label)
    tResult.evaluate(label)
tResult.display_result(num_evaluations)
```
